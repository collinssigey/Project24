```python
# Convert the columns to numeric and then to integers
df_movie_budgets['worldwide_gross'] = pd.to_numeric(df_movie_budgets['worldwide_gross'].str.replace(',', ''), errors='coerce').astype(int)
df_movie_budgets['production_budget'] = pd.to_numeric(df_movie_budgets['production_budget'].str.replace('$', ''), errors='coerce').astype(int)
```

```python
#create a new column in the DataFrame called 'Profit_Gross'
df_movie_budgets['Profit_Gross'] = df_movie_budgets['worldwide_gross'] - df_movie_budgets['production_budget']

#check the new DataFrame with the 'Profit_Gross' column
df_movie_budgets.tail()
```

```python
# Calculate the covariance between the time and gross profit
covariance = df_movie_budgets[['release_year', 'Profit_Gross']].cov().iloc[0, 1]

# Create a scatter plot of the data
plt.scatter(df_movie_budgets['release_year'], df_movie_budgets['Profit_Gross'])
plt.title('Covariance: {:.2f}'.format(covariance))
plt.xlabel('Release Year')
plt.ylabel('Gross Profit (Millions)')
plt.show()
```
